```python
import numpy as np
import matplotlib.pyplot as plt
import scipy
from scipy.interpolate import CubicSpline as spline
c=2.9979e10
G=6.67408e-8
Msun=1.989e33
Length = G*Msun/c**2
Time = Length/c
Density = Msun/Length**3
```

```python
m_r1=np.load('tid_def/SLy.npy')  
m_r2=np.load('tid_def/H4.npy')
m_r3=np.load('tid_def/ALF2.npy')
m_r4=np.load('tid_def/APR4.npy')
m_r5=np.load('tid_def/GNH3.npy')


k_l1=np.load('tid_def/k_l_SLy.npy')  
k_l2=np.load('tid_def/k_l_H4.npy')
k_l3=np.load('tid_def/k_l_ALF2.npy')
k_l4=np.load('tid_def/k_l_APR4.npy')
k_l5=np.load('tid_def/k_l_GNH3.npy')


```

```python
mx=np.amax(m_r1[0])
idx=np.where(m_r1[0]==mx)
idx=idx[0][0]
cs1=spline(m_r1[0][1:idx],k_l1[0][1:idx])
cs11=spline(m_r1[0][1:idx],m_r1[1][1:idx])

mx=np.amax(m_r2[0])
idx=np.where(m_r2[0]==mx)
idx=idx[0][0]
cs2=spline(m_r2[0][1:idx],k_l2[0][1:idx])
cs21=spline(m_r2[0][1:idx],m_r2[1][1:idx])

mx=np.amax(m_r3[0])
idx=np.where(m_r3[0]==mx)
idx=idx[0][0]
cs3=spline(m_r3[0][1:idx],k_l3[0][1:idx])
cs31=spline(m_r3[0][1:idx],m_r3[1][1:idx])

mx=np.amax(m_r4[0])
idx=np.where(m_r4[0]==mx)
idx=idx[0][0]
cs4=spline(m_r4[0][1:idx],k_l4[0][1:idx])
cs41=spline(m_r4[0][1:idx],m_r4[1][1:idx])

mx=np.amax(m_r5[0])
idx=np.where(m_r5[0]==mx)
idx=idx[0][0]
cs5=spline(m_r5[0][1:idx],k_l5[0][1:idx])
cs51=spline(m_r5[0][1:idx],m_r5[1][1:idx])

```

```python
#SLy
m1=np.zeros(7)
r1=np.zeros(7)
k1=np.zeros(7)
mass=1.2
i=0
while mass<1.36:
    m1[i]=mass
    r1[i]=cs11(mass)*Length/1.0e5
    k1[i]=cs1(mass)
    i=i+1
    mass=mass+0.025
    
print('SLy EOS')
print(' Mass   Radius     k')
for i in range(7):
    print('%.3f' %m1[i],'|','%.3f' %r1[i],'|','%.4f' %k1[i])
    
    
    
#H4
m2=np.zeros(7)
r2=np.zeros(7)
k2=np.zeros(7)
mass=1.2
i=0
while mass<1.36:
    m2[i]=mass
    r2[i]=cs21(mass)*Length/1.0e5
    k2[i]=cs2(mass)
    i=i+1
    mass=mass+0.025

print()
print('H4 EOS')
print(' Mass   Radius     k')
for i in range(7):
    print('%.3f' %m2[i],'|','%.3f' %r2[i],'|','%.4f' %k2[i])
    
#ALF2
m3=np.zeros(7)
r3=np.zeros(7)
k3=np.zeros(7)
mass=1.2
i=0
while mass<1.36:
    m3[i]=mass
    r3[i]=cs31(mass)*Length/1.0e5
    k3[i]=cs3(mass)
    i=i+1
    mass=mass+0.025

print()
print('ALF2 EOS')
print(' Mass   Radius     k')
for i in range(7):
    print('%.3f' %m3[i],'|','%.3f' %r3[i],'|','%.4f' %k3[i])
    
    
    
#APR4
m4=np.zeros(7)
r4=np.zeros(7)
k4=np.zeros(7)
mass=1.2
i=0
while mass<1.36:
    m4[i]=mass
    r4[i]=cs41(mass)*Length/1.0e5
    k4[i]=cs4(mass)
    i=i+1
    mass=mass+0.025

print()
print('APR4 EOS')
print(' Mass   Radius     k')
for i in range(7):
    print('%.3f' %m4[i],'|','%.3f' %r4[i],'|','%.4f' %k4[i])
    
    
#GNH3
m5=np.zeros(7)
r5=np.zeros(7)
k5=np.zeros(7)
mass=1.2
i=0
while mass<1.36:
    m5[i]=mass
    r5[i]=cs51(mass)*Length/1.0e5
    k5[i]=cs5(mass)
    i=i+1
    mass=mass+0.025

print()
print('GNH3 EOS')
print(' Mass   Radius     k')
for i in range(7):
    print('%.3f' %m5[i],'|','%.3f' %r5[i],'|','%.4f' %k5[i])
```
